### State Wise Analysis of Sales, Discount and Profit

 - Vermont at the Top
 - Ohio at the Bottom

```python
df_state.sort_values('Discount')[[ 'Discount']].plot(kind = 'bar', figsize = (15,10))
plt.title('State wise Discount Analysis', fontsize = 20)
plt.ylabel('Discount per Sate')
plt.xlabel('States')
plt.show()
```
